```python
# Karatsuba's Algorithm

def Karatsuba(a, b):
    # Base case
    if len(a) == 1 or len(b) == 1:
        return str(int(a) * int(b))
    
    # Find right cut-off (from the right, choose the shorter one)
    m = min(len(a) // 2, len(b) // 2)
    
    # Slice input strings to parts
    x1 = a[0:len(a)-m]
    x0 = a[len(a)-m:]
    y1 = b[0:len(b)-m]
    y0 = b[len(b)-m:]
    
    # Calculate smaller products
    z2 = int(Karatsuba(x1, y1))
    z1 = int(Karatsuba(x0, y1)) + int(Karatsuba(x1, y0))
    z0 = int(Karatsuba(x0, y0))
    
    # Combine products and return as string
    return str(10**(2*m) * z2 + 10**(m) * z1 + z0)

Karatsuba(a, b)
```

```python
def DaCInversionCount(arr):
    # Base case
    if len(arr) <= 1:
        return 0
    
    # Cut array at the center, and assume left and right are sorted
    cut = len(arr) // 2
    leftarr = sorted(arr[0:cut])
    rightarr = sorted(arr[cut:])
    
    # Everytime we get a value from the right aray, add # of remaining element in the left array to the counter
    leftpointer = 0
    rightpointer = 0
    mergecount = 0
    while leftpointer < len(leftarr) and rightpointer < len(rightarr):
        if leftarr[leftpointer] > rightarr[rightpointer]:
            mergecount += len(leftarr) - leftpointer
            rightpointer += 1
        else:
            leftpointer += 1
    
    # Recursive call
    return DaCInversionCount(arr[0:cut]) + DaCInversionCount(arr[cut:]) + mergecount

DaCInversionCount(data)
```

```python
def QuickSort1Pivot(arr, start, end):
    # Base case
    if end - start <= 0:
        return (arr, 0)
    
    # Set pivot and </> pivot split position
    pivot = arr[start]
    split = start + 1
    
    # Flip position only when new element is < pivot, switch with element at split position
    compcnt = 0
    for i in range(start + 1, end + 1):
        compcnt += 1
        if arr[i] < pivot:
            arr[i], arr[split] = arr[split], arr[i]
            split += 1
            
    # Switch pivot with split
    arr[split-1], arr[start] = arr[start], arr[split-1]
    
    # Recursive call to handle left and right of pivot
    (arr, leftres) = QuickSort1Pivot(arr, start, split-2)
    (arr, rightres) = QuickSort1Pivot(arr, split, end)
    return (arr, leftres + rightres + compcnt)
    
QuickSort1Pivot(data3,0,len(data3)-1)
```

```python
from typing import List, Tuple
from collections import defaultdict
import random
import copy


def build_graph(rows):
    graph = defaultdict(lambda: {'fused': [], 'connected': {}})

    for row in rows:
        this_node, connected = row[0], row[1:]

        for n in connected:
            graph[this_node]['connected'][n] = True
            graph[n]['connected'][this_node] = True
  
    return graph


def pick_edge(graph: dict):
    from_node = random.choice(list(graph.keys()))
    to_node = random.choice(list(graph[from_node]['connected'].keys()))
    return from_node, to_node


def fuse_edge(graph: dict, edge: Tuple[int, int]):

    del graph[edge[0]]['connected'][edge[1]]
    del graph[edge[1]]['connected'][edge[0]]
    
    clean_up_edges = []
    combined_node = edge[0]
    delete_node = edge[1]

    for connected_to in graph[delete_node]['connected']:
        if combined_node != connected_to:
            graph[combined_node]['connected'][connected_to] = True
            graph[connected_to]['connected'][combined_node] = True
            clean_up_edges.append((connected_to, delete_node))

    graph[combined_node]['fused'].append(delete_node)
    graph[combined_node]['fused'].extend(graph[delete_node]['fused'])
    del graph[delete_node]

    for stale_edge in clean_up_edges:
        del graph[stale_edge[0]]['connected'][stale_edge[1]]
    
    return graph


def random_contract(graph: dict):
    while len(graph.items()) > 2:
        edge: Tuple[int, int] = pick_edge(graph)
        graph = fuse_edge(graph, edge)

    partitions = []

    for node in graph:
        partitions.append(sorted([node] + graph[node]['fused']))

    return tuple(partitions)


def count_cuts(partitions: Tuple[List[int], List[int]], graph: dict):
    total = 0
    a, b = partitions

    for i in a:
        for j in b:
            if graph[i]['connected'].get(j):
                total += 1

    return total

graph = build_graph(data4)
best_count = None

for x in range(0, 20000):
    partitions = random_contract(copy.deepcopy(graph))
    cut_count = count_cuts(partitions, graph)

    if best_count is None or cut_count < best_count:
        best_count = cut_count
        print('New best min cut: ' + str(best_count))
    if x % 1000 == 0:
        print("Round: " + str(x))
        
print('Best: %s' % best_count)
```
