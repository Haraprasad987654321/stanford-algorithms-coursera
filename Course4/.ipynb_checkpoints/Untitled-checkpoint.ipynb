# Programming Exercise 2

```python
"""
Traveling salesman problem
"""

import numpy as np
from itertools import combinations

f = open('C4W2_tsp.txt', 'r')
ls = f.readlines()[1:]
graph = [list(map(float, i.split(' '))) for i in ls]


def dis(i, j):
    return np.sqrt((graph[i][0]-graph[j][0])**2+(graph[i][1]-graph[j][1])**2)


N = len(graph)
dic1 = {frozenset([0]): {0: 0}}

for m in range(1, N):
    print(m)
    comb = list(combinations(range(1, N), m))
    dic2 = {frozenset(comb[i]): {list(comb[i])[j]: 0 for j in range(m)} for i in range(len(comb))}
    print(m, len(dic2))
    for s in dic2:
        for j in s:
            ans = []
            if m == 1:
                dic2[s][j] = dis(0, j)
            else:
                sj = set(s)
                sj.remove(j)
                dic2[s][j] = min([dic1[frozenset(sj)][k]+dis(k, j) for k in sj if k != j])
    dic1 = dic2.copy()

tsp = min([dic2[frozenset(comb[0])][j]+dis(0, j) for j in range(1, N)])
print(tsp)
```

```python
text_file = open("C4W2_tsp.txt", "r")
lines = text_file.readlines()
data = [i.strip().split(" ") for i in lines[0:]]
data = [[float(j) for j in i] for i in data]
n = int(data[0][0])
cities = data[1:]
text_file.close()
```

```python
x = []
y = []
for i, j in cities:
    x.append(i)
    y.append(j)
import matplotlib.pyplot as plt
plt.plot(x, y)

# zip joins x and y coordinates in pairs
for i in range(n):

    label = i

    plt.annotate(label, # this is the text
                 (x[i],y[i]), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.show()
```

```python
import numpy as np
from itertools import combinations

distdict = {}
comb = list(combinations(range(0, n), 2))
for i, j in comb:
    distdict[i, j] = np.sqrt((cities[i][0]-cities[j][0])**2+(cities[i][1]-cities[j][1])**2)

def tot(seq):
    tot = 0
    for i in range(0,25):
        if (seq[i-1], seq[i]) in distdict.keys():
            tot += distdict[seq[i-1], seq[i]]
        else:
            tot += distdict[seq[i], seq[i-1]]
    return tot
    
```

```python
seq = [0,1,5,9,10,11,14,16,17,18,20,21,22,24,23,19,15,13,12,8,7,6,2,3,4]
minres = tot(seq)
print(minres)
for i in range(len(seq)-1):
    for j in range(i+1, len(seq)):
        seq = [0,1,5,9,10,11,14,16,17,18,20,21,22,24,23,19,15,13,12,8,7,6,2,3,4]
        seq[i], seq[j] = seq[j], seq[i]
        res = tot(seq)
        if res < minres:
            minres = res
            print(seq)
            print(minres)
seq = [0, 1, 5, 9, 10, 11, 14, 16, 17, 18, 20, 21, 22, 19, 23, 24, 15, 13, 12, 8, 7, 6, 2, 3, 4]
minres = 29357.197606509893   
for i in range(len(seq)-2):
    for j in range(i+1, len(seq)-1):
        for k in range(j+1, len(seq)):
            seq[i], seq[j], seq[k] = seq[j], seq[k], seq[i]
            res = tot(seq)
            if res < minres:
                minres = res
                print(seq)
                print(minres)
            seq[i], seq[j], seq[k] = seq[k], seq[i], seq[j]
            res = tot(seq)
            if res < minres:
                minres = res
                print(seq)
                print(minres)

seq = [0, 1, 5, 9, 10, 11, 14, 18, 17, 22, 20, 21, 16, 19, 23, 24, 15, 13, 12, 8, 7, 6, 2, 3, 4]
minres = 28146.989794246838 
for i in range(len(seq)-3):
    for j in range(i+1, len(seq)-2):
        for k in range(j+1, len(seq)-1):
            for l in range(k+1, len(seq)):
                seq[i], seq[j], seq[k], seq[l] = seq[j], seq[k], seq[l], seq[i]
                res = tot(seq)
                if res < minres:
                    minres = res
                    print(seq)
                    print(minres)
                seq[i], seq[j], seq[k], seq[l] = seq[l], seq[i], seq[j], seq[k]
                res = tot(seq)
                if res < minres:
                    minres = res
                    print(seq)
                    print(minres)
```
